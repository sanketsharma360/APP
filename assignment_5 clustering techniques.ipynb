```python
x = df.iloc[:,3:] # add two columns in x
```

```python
plt.title('Unclustered data')
plt.xlabel('Annual Income ')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)']) #this is a ublabeld data
```

```python
#
#elbow method is visualization method 
```

```python
from sklearn.cluster import KMeans , AgglomerativeClustering
```

```python
km = KMeans(n_clusters=3) # to create the number of clustered you want to create
```

```python
x.shape # there are 200 customers by using fit_predict method it gives the label to the 200 customers
```

```python
km.fit_predict(x) # it gives the clusters label 
```

```python
# sum squred error Inertia measures how well a dataset was clustered by K-Means.
km.inertia_
```

```python
sse = []
for k in range(1,16):
    km = KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title('Elbow Method')
plt.xlabel('Value of K')
plt.ylabel('SSE')
plt.grid()
plt.xticks(range(1,16)) # vertical lines are 16
plt.plot(range(1,16),sse,marker='.',color='red') # ideal value of the k is 5k
```

```python
km = KMeans(n_clusters=5,random_state=0)
```

```python
label = km.fit_predict(x)
```

```python
label
```

```python
km.cluster_centers_
```

```python
centroids =  km.cluster_centers_
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1) #  1 row , 2 col , 1st plot
plt.title('Unclustered data')
plt.xlabel('Annual Income ')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'])

plt.subplot(1,2,2)  #  1 row , 2 col , 2st plot
plt.title('clustered data')
plt.xlabel('Annual Income ')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'] , c= label) 
plt.scatter(centroids[:,0], centroids[:,1] , s=100 , color ='k')
```
